8.	Apply an appropriate classification algorithm to build a model. Compare various models with the results from KNN.

```python
features = new_df.iloc[:,[0,1,2,3,4,5,6,7]].values
label = new_df.iloc[:,8].values
```

```python
#Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,label,test_size=0.2,random_state =10)
```

```python
# Logestic Regression
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train) 
```

```python
# Accuracy of the train data set by Logistic Regression
lr_model.score(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
y_pred
```

```python
# Accuracy of the test data set by Logistic Regression
lr_model.score(x_test, y_test)
```

```python
#Preparing ROC Curve (Receiver Operating Characteristics Curve)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# predict probabilities
probs = lr_model.predict_proba(features)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(label, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(label, probs)
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.plot(fpr, tpr, marker='.')
```

```python
model = []
model_score = []
model_auc = []
model.append("LR")
model_score.append(lr_model.score(x_test, y_test))
model_auc.append(auc)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
y_pred_rfc = rfc.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_rfc))
```

```python
confusion = confusion_matrix(y_test, rfc.predict(x_test))
```

```python
confusion
```

```python
# True Positive
TP = confusion[1,1]

# True Negatives
TN = confusion[0,0] 

# False Positives
FP = confusion[0,1]

# False Negatives
FN = confusion[1,0] 
```

```python
Accuracy = (TP+TN)/(TP+TN+FP+FN)
Precision = TP/(TP+FP)
Sensitivity = TP/(TP+FN)                  
Specificity = TN/(TN+FP)
```

```python
print("Accuracy:", Accuracy)
print("Precision: %.2f" %Precision)
print("Sensitivity: %.2f" %Sensitivity)
print("Specificity: %.2f" %Specificity)
print("AUC: %.2f" %auc_rfc)
```
